```python
import pandas as pd
import numpy as np
```

```python
with open('Input.txt') as f:
    lines = f.readlines()
```

# Part 1

```python
for idx, l in enumerate(lines): 
    lines[idx] = l.replace("\n", "")
```

```python
lines
```

```python
lines = [list(l) for l in lines]
```

```python
df = pd.DataFrame(lines)
df = df.astype("int")
```

```python
gamma = []
epsilon = []
for i in range(len(lines[0])):
    if df.apply(sum)[i] > 500:
        gamma.extend([1])
        epsilon.extend([0])
    else:
        gamma.extend([0])
        epsilon.extend([1])
    
```

```python
print(gamma, epsilon)
```

```python
(int("".join([str(g) for g in gamma]),2)) * (int("".join([str(g) for g in epsilon]),2))
```
